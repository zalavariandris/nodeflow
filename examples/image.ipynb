```python
import numpy as np
import matplotlib.pyplot as plt
import nodeflow as nf
```

```python
ramp = nf.image.Ramp( (512,512) )
img = nf.evaluate(ramp)
plt.imshow(img)
```

```python
from nodeflow.core import Constant
url = nf.Constant("https://upload.wikimedia.org/wikipedia/commons/5/5c/Burosch_Blue-Only_Test_pattern.jpg")
read = nf.image.Read(url)

img = nf.evaluate(read)
print(img)
```
